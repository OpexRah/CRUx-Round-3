```python
def print_categorymembers(categorymembers, level=0, max_level=1):
        flag = False
        for c in categorymembers.values():
            print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
            if c.ns == 0:
              cleaned_summary = ' '.join(c.summary.split())
              titles.append(c.title)
              summary.append(c.summary)
              if len(summary) >= 400:
                return
            if c.ns == wapi.Namespace.CATEGORY and level < max_level:
                print_categorymembers(c.categorymembers, level=level + 1, max_level=max_level)

titles = []
summary = []
cat = test.page("Category:Cities")
print("Category members: Category:Cities")
print_categorymembers(cat.categorymembers)
print(len(titles), len(summary))
```

Trying to use the special URL for random articles:
URL:  https://en.wikipedia.org/wiki/Special:Random

```python
import requests
from bs4 import BeautifulSoup


r_title = []
r_summary = []
i = 0
while len(r_summary) < 200:
  url = requests.get("https://en.wikipedia.org/wiki/Special:Random")

  soup = BeautifulSoup(url.content, "html.parser")
  p_title = soup.find(class_="firstHeading").text

  page = test.page(p_title)

  if page.exists():
    r_title.append(page.title)
    summary = page.summary
    print("Item number ", i)

    cleaned_summary = ' '.join(summary.split())

    r_summary.append(cleaned_summary)
    i += 1
  else:
    pass

print(len(r_title), len(r_summary))
```

```python
import re
#cleaning the data
for x in range(0, len(r_summary)):
  final_summary = ""
  word_list = r_summary[x].split(" ")
  for word in word_list:
      #word_filtered = ''.join(filter(str.isalnum, word))
      pattern = r'[^A-Za-z0-9]+'
      word_filtered = re.sub(pattern, '', word)
      final_summary += word_filtered + " "
  r_summary[x] = final_summary
```

```python
import csv
data_file = open('./random articles cleaned.csv', 'w', newline='')
csv_writer = csv.writer(data_file)
csv_writer.writerow(["title", "summary"])

for i in range(0, len(r_title)):
  new = [r_title[i], r_summary[i]]
  csv_writer.writerow(new)

data_file.close()
```
